# EDA

## Name Column

```python
plt.figure(figsize=(12,6))
df.name.value_counts()[:20].plot(kind='barh')
```

**Observation**
- **Maruti Swift Dzire VDI** is best suitable car

 ## Year Column

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df,x=df.year,bins=40)
plt.title("Distribution of Year",fontsize=20)
plt.figure(figsize=(12,6))
df.year.value_counts().plot(kind='bar')
plt.figure(figsize=(12,6))
plt.title("Distribution of Year Wise Top 10 Model",fontsize=20)
df.year.value_counts()[:10].plot(kind='pie',autopct='%.1f%%')
plt.legend(loc=6)
plt.show()
```

**Observation**
- **2017 Model Car** has highest ratio in dataset

## Fuel Columns

```python
plt.figure(figsize=(12,6))
plt.title("Fuel Type Distribution",fontsize=20)
sns.countplot(data=df,x=df.fuel)
plt.show()
plt.figure(figsize=(12,6))
df.fuel.value_counts().plot(kind='pie',autopct='%.1f%%')
plt.legend()
plt.show()
```

```python
pd.DataFrame(df.fuel.value_counts()/len(df)*100)
```

**Observation**
- The Most of the Car is **fuel type** is **Diesel and Petrol** about **54% and 44 %** Respectively.
- other fuel type are **CNG and LPG** which is **less than 1%.**

## Seller Type Column

```python
plt.figure(figsize=(12,6))
plt.title("Seller Type",fontsize=20)
sns.countplot(data=df,x=df.seller_type)
plt.show()
plt.figure(figsize=(12,6))
plt.title("Seller Type",fontsize=20)

df.seller_type.value_counts().plot(kind='pie',autopct='%.1f%%')
plt.legend(loc=4)
plt.show()
```

**Observation**
- Most of the car sale by the **Individual person** own car **ratio is 83%.**
- Most of the **Dealer** sale the car those **ratio is 14%.**
- Most of the **Trustmark-dealer** sale the car those **ratio is 3%.**

## Transmission

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.countplot(data=df,x=df.transmission)
plt.title("Transmission Mod of car",fontsize=20)
plt.subplot(1,2,2)
plt.title("Transmission Mod of car",fontsize=20)
df.transmission.value_counts().plot(kind='pie',autopct='%.1f%%')
plt.legend(loc=1)
plt.show()
```

**Observation**
- Most of the car **transmission mode** is **Manual.**

## Owner Column

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x=df.owner)
plt.title("Owner of the Car",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Owner of the Car",fontsize=20)
df.owner.value_counts().plot(kind='pie',autopct='%.1f%%')
plt.legend(loc=1)
plt.show()
```

**Observation**
- Maximum number of the car sale by the **First_Owner** ratio is **65%.**
- **26% car** sale by the **Second_Owner.**
- **7% car** sale b the the **Third-Owner.**

## Seats Column

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.countplot(data=df,x=df.seats)
plt.title("Seats Capacity of Car",fontsize=20)
plt.subplot(1,2,2)
df.seats.value_counts().plot(kind='pie',autopct='%.1f%%')
plt.title("Seats Capacity of Car",fontsize=20)
plt.legend(loc=4)
plt.show()
```

**Observation**
- The number of car has **5 seats** are available for sale ratio is high

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Diesel VS Price",fontsize=20)
sns.histplot(data=df,x=df[df['fuel']=='Diesel']['selling_price'],kde=True,bins=25)
plt.subplot(1,2,2)
plt.title("Petrol VS Price",fontsize=20)

sns.histplot(data=df,x=df[df['fuel']=='Petrol']['selling_price'],kde=True,bins=25)
```

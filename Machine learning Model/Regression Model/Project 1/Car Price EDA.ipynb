# EDA

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df,x=df.year,bins=40)
plt.title("Distribution of Year",fontsize=20)
plt.figure(figsize=(12,6))
df.year.value_counts().plot(kind='bar')
```

**Observation**
- **2017 Model Car** has highest ratio in dataset

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.countplot(data=df,x=df.fuel)
plt.subplot(1,2,2)
df.fuel.value_counts().plot(kind='pie',autopct="")
```

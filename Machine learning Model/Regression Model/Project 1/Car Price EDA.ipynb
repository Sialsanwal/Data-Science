# EDA

## Year Column

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df,x=df.year,bins=40)
plt.title("Distribution of Year",fontsize=20)
plt.figure(figsize=(12,6))
df.year.value_counts().plot(kind='bar')
plt.figure(figsize=(12,6))
plt.title("Distribution of Year Wise Top 10 Model",fontsize=20)
df.year.value_counts()[:10].plot(kind='pie',autopct='%.1f%%')
plt.legend(loc=6)
plt.show()
```

**Observation**
- **2017 Model Car** has highest ratio in dataset

```python
plt.figure(figsize=(12,6))
plt.title("Fuel Type Distribution",fontsize=20)
sns.countplot(data=df,x=df.fuel)
plt.show()
plt.figure(figsize=(12,6))
df.fuel.value_counts().plot(kind='pie',autopct='%.1f%%')
plt.legend()
plt.show()
```

```python
pd.DataFrame(df.fuel.value_counts()/len(df)*100)
```

**Observation**
- The Most of the Car is **fuel type** is **Diesel and Petrol** about **54% and 44 %** Respectively.
- other fuel type are **CNG and LPG** which is **less than 1%.**

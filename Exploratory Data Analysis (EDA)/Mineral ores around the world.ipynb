# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

# Load Dataset

```python
df=pd.read_csv('Data Sets/Mineral ores round the world.csv')
```

```python
df.head()
```

## Region which most Mineral Sites

```python
sns.countplot(data=df,x=df['region'])
```

**Observation**
- The South Austrilla has maximum Mineral sites

## Country with most Mineral Site

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x=df['country'].head(5000))
plt.xticks(rotation=90)
plt.show()
```

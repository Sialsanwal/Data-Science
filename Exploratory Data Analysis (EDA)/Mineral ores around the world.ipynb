# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

# Load Dataset

```python
df=pd.read_csv('Data Sets/Mineral ores round the world.csv')
```

```python
df.head()
```

```python
df.shape
```

## Region which most Mineral Sites

```python
sns.countplot(data=df,x=df['region'])
```

**Observation**
- The South Austrilla has maximum Mineral sites

## Country with most Mineral Site

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x=df['country'].head(5000))
plt.xticks(rotation=90)
plt.show()
```

**Observation**
- The **United States** has maximum number of mineral sites

## State Wise with most Mineral Site

```python
plt.figure(figsize=(15,8))
sns.countplot(data=df,x=df['state'].head(5000))
plt.xticks(rotation=90)
plt.show()
```

**Observation**
- The **Alaska State** has maximum Mineral Site

## Com Type

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x=df['com_type'])
plt.title("Command Type of Mineral Site",fontsize=20)
plt.show()
```

**Observation**
- The Command **M** has maximum mineral

## Commod1

```python
plt.figure(figsize=(20,7))
sns.countplot(data=df,x=df['commod1'].head(500))
plt.title("Command Type of Mineral Site",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

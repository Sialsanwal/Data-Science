# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

# Load Dataset

```python
df=pd.read_csv('Data Sets/Mineral ores round the world.csv')
```

```python
df.head()
```

## Check the Shape of data

```python
df.shape
```

**Observation**

## Region which most Mineral Sites

```python
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
sns.countplot(data=df,x=df['region'])
plt.title("Region which most Mineral Sites",fontsize=15)

plt.subplot(1,2,2)
df.region.value_counts().plot(kind='pie',autopct='%.1f%%')
#plt.figure(figsize=(12,6))
plt.title("Region which most Mineral Sites",fontsize=15)
plt.legend(loc=0)
```

**Observation**
- The **South Austrilla** has maximum Mineral sites about **53%**

## Country with most Mineral Site

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x=df['country'].head(5000))
plt.xticks(rotation=90)
plt.title("Country with most Mineral Site",fontsize=20)
plt.show()
```

**Observation**
- The **United States** has maximum number of mineral sites

## State Wise with most Mineral Site

```python
plt.figure(figsize=(15,8))
sns.countplot(data=df,x=df['state'].head(5000))
plt.xticks(rotation=90)
plt.title("State Wise with most Mineral Site",fontsize=20)
plt.show()
```

**Observation**
- The **Alaska State** has maximum Mineral Site

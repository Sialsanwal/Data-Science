# 1. Matplotlib


# 1.1. Line Chart


```python
x=np.linspace(-1,1,50)
y1=2*x+1
plt.figure()
plt.plot(x,y1)

plt.xlabel("Its X-axis")
plt.ylabel("Its Y-aixs")
plt.title("Line Chart")
```

```python
Year = [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
Unemployment_Rate = [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
plt.plot(Year,Unemployment_Rate)
```

# 1.2. Bar Chart

```python
data={"C":20,
     "C++":15,
     "Java":30,
     "python":35}
courses=list(data.keys())
values=list(data.values())
plt.bar(courses,values,color='yellow')
plt.title("Student Enroll in different courses")
plt.xlabel("Course Offer")
plt.ylabel("No of student enroll")
```

# 2.1. Line Chart


```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights.shape
```

```python
flights["month"].unique()
```

```python
may_flight=flights.query("month=='May'")
sns.lineplot(data=may_flight,x="year",y="passengers")
```

```python
sns.barplot(data=may_flight,x="year",y="passengers")
```

```python
sns.barplot(data=flights,x='year',y='passengers',hue='month')
```

```python
flight_wide=flights.pivot("year",'month','passengers')
```

```python
flight_wide.head()
```

```python
flights.pivot(index='month',columns='year',values='passengers').head()
```

```python
sns.lineplot(data=flight_wide)
```

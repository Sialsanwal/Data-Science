- 4 Write a Python program to get the difference between a given number and 17, if the number is greater than 17 return double the absolute difference.

```python
def numdiff():
    a=int(input("Enter a number:"))
    d=a-17
    if d>17:
        s=2*d
        return s
    else:
        return d
```

```python
numdiff()
```

 - 5 Write a Python program to calculate the sum of three given numbers, if the values are equal then return three times their sum.

```python
def sum_three(x,y,z):
    sum= x+y+z
    if x==y==z:
        sum=sum*3
    return sum
```

```python
sum_three(5,5,5)
```

 - 6 Write a Python program to test whether a passed letter is a vowel or not.


```python
vowels=['a','e','i','o','u','A','E','I','O','U']
def vowel_alphabet(str):
    if str in vowels:
        print("The given alphabet is {} is vowels".format(str))
    else:
        print("The given alphabet is {} is constanents".format(str))
        
```

```python
vowel_alphabet("h")
```

```python
vowel_alphabet("O")
```

```python
users=str(input("Enter the Alphabets:"))
if users in vowels:
    print("The given alphabet is {} is vowels".format(users))
else:
    print("The given alphabet is {} is constanents".format(users))
        
```

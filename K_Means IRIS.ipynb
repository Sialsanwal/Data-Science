```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
from sklearn.metrics import davies_bouldin_score, silhouette_score, silhouette_samples
from sklearn.preprocessing import StandardScaler


import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("D:/Data-Science/Machine learning Model/Classification/DataSet/iris.csv")
```

```python
df.head()
```

```python
df.iloc[:,:4] = StandardScaler().fit_transform(df.iloc[:,:4])
# Exploratory Data Analysis
sns.pairplot(df,diag_kind='kde')
sns.pairplot(df, hue="variety", diag_kind= 'kde') 
```

**Observation**
- clearly show that 3 cluster are formed

```python
sse=[]
k_rng=range(1,10)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['sepal.length',"sepal.width","petal.length","petal.width"]])
    sse.append(km.inertia_)
```

```python
plt.xlabel("K")
plt.ylabel("Sum of Squared Error")
plt.plot(k_rng,sse)
```

**Observation**
- the elbow point is 3
